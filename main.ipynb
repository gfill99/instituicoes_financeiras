```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import functions as fc
```

```python
ano = 2017
ano_mes = 2019
ano_final = 2022
```

```python
lista_tickers = fc.list_tickers
lista_tickers
```

```python
acao1_anomes = fc.tratar_dados(fc.carregar_dados(lista_tickers[0], ano_mes, ano_final), 'ano_mes')
acao2_anomes = fc.tratar_dados(fc.carregar_dados(lista_tickers[1], ano_mes, ano_final), 'ano_mes')
acao3_anomes = fc.tratar_dados(fc.carregar_dados(lista_tickers[2], ano_mes, ano_final), 'ano_mes')
acao4_anomes = fc.tratar_dados(fc.carregar_dados(lista_tickers[3], ano_mes, ano_final), 'ano_mes')
acao5_anomes = fc.tratar_dados(fc.carregar_dados(lista_tickers[4], ano_mes, ano_final), 'ano_mes')
```

```python
acao1_ano = fc.tratar_dados(fc.carregar_dados(lista_tickers[0], ano, ano_final), 'ano')
acao2_ano = fc.tratar_dados(fc.carregar_dados(lista_tickers[1], ano, ano_final), 'ano')
acao3_ano = fc.tratar_dados(fc.carregar_dados(lista_tickers[2], ano, ano_final), 'ano')
acao4_ano = fc.tratar_dados(fc.carregar_dados(lista_tickers[3], ano, ano_final), 'ano')
acao5_ano = fc.tratar_dados(fc.carregar_dados(lista_tickers[4], ano, ano_final), 'ano')
```

```python
# Gráfico do retorno logarítmico
plt.figure(figsize=(14, 8))  # Ajustando o tamanho da figura

# Plotando os dados
plt.plot(acao1_anomes['ano_mes'], acao1_anomes['log_return'], label=lista_tickers[0].replace('.SA', ''), color='orange', linewidth=2)
plt.plot(acao2_anomes['ano_mes'], acao2_anomes['log_return'], label=lista_tickers[1].replace('.SA', ''), color='red', linewidth=2)
plt.plot(acao3_anomes['ano_mes'], acao3_anomes['log_return'], label=lista_tickers[2].replace('.SA', ''), color='yellow', linewidth=2)
plt.plot(acao4_anomes['ano_mes'], acao4_anomes['log_return'], label=lista_tickers[3].replace('.SA', ''), color='blue', linewidth=2)
plt.plot(acao5_anomes['ano_mes'], acao5_anomes['log_return'], label=lista_tickers[4].replace('.SA', ''), color='black', linewidth=2)

# Título e Labels
plt.title('Retorno médio das Instituições Financeiras', fontsize=16)
plt.xlabel('Ano-Mês', fontsize=12)
plt.ylabel('Retorno Logarítmico (%)', fontsize=12)

# Ajustar a formatação das datas no eixo X
plt.xticks(rotation=45, ha='right', fontsize=10)

# Adicionando o grid para melhorar a leitura
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adicionando legenda
plt.legend(loc='upper left', fontsize=12)

# Ajustando o layout para que nada se sobreponha
plt.tight_layout()

# Mostrar gráfico
plt.show()
```
